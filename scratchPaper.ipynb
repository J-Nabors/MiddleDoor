```python
import requests
import os

# Census API endpoint
BASE_URL = "https://api.census.gov/data"

# You need to get a Census API key from https://api.census.gov/data/key_signup.html
API_KEY = "YOUR_API_KEY_HERE"  # Replace with your actual API key

# Download ACS data for each table
for table in tables:
    for year in range(2011, 2024):
        # Construct the API URL for ACS 5-Year estimates
        url = f"{BASE_URL}/{year}/acs/acs5"
        
        # Get all variables for the table
        params = {
            "get": f"group({table})",
            "for": "zip code tabulation area:*",
            "key": API_KEY
        }
        
        try:
            response = requests.get(url, params=params)
            response.raise_for_status()
            
            # Save to CSV
            output_file = os.path.join(folder_path, f"{table}_{year}_acs5.csv")
            
            # Write response data to file
            with open(output_file, 'w') as f:
                f.write(response.text)
            
            print(f"Downloaded {table} for {year}")
        
        except requests.exceptions.RequestException as e:
            print(f"Error downloading {table} for {year}: {e}")
```
